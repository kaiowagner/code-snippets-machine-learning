# Multiclass Classification

```python
import pandas as pd

dataset = pd.read_csv('../../../datasets/iris.csv')
dataset.head()
```

## Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils

data = dataset.iloc[:, 0:4].values
target = dataset.iloc[:,4].values

label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)

target_dummy = np_utils.to_categorical(target)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target_dummy, test_size=0.25)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# units = Quantos neurônios farão parte da camada oculta
# Para determinar o número de neurônios: (número de atributos previsores + número de classes)/ 2
# activation: função de ativação
# kernel_initializer: Incialização dos pesos
# input_dim = Quantos elementos na camada de entrada

classifier = Sequential()
# Camada oculta
classifier.add(Dense(units=4, activation='relu', input_dim=4))
classifier.add(Dense(units=4, activation='relu'))

# Camada de saída 
# Usamos a sigmoid, pois é um problema de classificação com mais de duas classes
classifier.add(Dense(units=3, activation='softmax'))

# optimizer: Função para ajuste dos pesos (descida do gradiente...)
# loss: Função de perda. Como o erro será calculado
# metrics: Qual a métrica que será utilizada para avaliação
classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
# Calcular o erro para "batch_size" e depois atualizar os pesos
# epochs: Quantas épocas eu quero executar o treinamento
classifier.fit(X_train, y_train, batch_size = 10, epochs=1000)
```

## Predictions

```python
from sklearn.metrics import confusion_matrix
import numpy as np

result = classifier.evaluate(X_test, y_test)
print(result)

predictions = classifier.predict(X_test)
predictions = (predictions > 0.5)

# Retorna o indice de maior valor
y_test_2 = [np.argmax(t) for t in y_test]
predictions_2 = [np.argmax(t) for t in predictions]


precision = confusion_matrix(y_test_2, predictions_2)
print(precision)
```

## Cross Validation

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
import keras
```

```python
def build_network():
    classifier = Sequential()
    classifier.add(Dense(units=4, activation='relu', input_dim=4))
    classifier.add(Dense(units=4, activation='relu'))
    classifier.add(Dense(units=3, activation='softmax'))
    classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
    return classifier
```

```python
classifier = KerasClassifier(build_fn=build_network, epochs=1000, batch_size=10)
```

```python
results = cross_val_score(estimator=classifier, X=data, y=target, cv=10, scoring='accuracy')
```
