## Label Encoder

```python
data = base.iloc[:, 1:13].values
price_target = base.iloc[:,0].values
```

```python
from sklearn.preprocessing import LabelEncoder

labelenconder_data = LabelEncoder()
data[:, 0] = labelenconder_data.fit_transform(data[:, 0])
data[:, 1] = labelenconder_data.fit_transform(data[:, 1])
data[:, 3] = labelenconder_data.fit_transform(data[:, 3])
data[:, 5] = labelenconder_data.fit_transform(data[:, 5])
data[:, 8] = labelenconder_data.fit_transform(data[:, 8])
data[:, 9] = labelenconder_data.fit_transform(data[:, 9])
data[:, 10] = labelenconder_data.fit_transform(data[:, 10])

data
```

## One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder(categorical_features = [0,1,3,5,8,9,10])
data = one_hot_encoder.fit_transform(data).toarray()
```

```python
data
```

## Train

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
regressor = Sequential()
# Para determinar o número de neurônios: (número de atributos previsores + número de classes)/ 2
# Utiliza-se 'relu' para problemas de regressão
regressor.add(Dense(units=158, activation='relu',input_dim=323))
regressor.add(Dense(units=158, activation='relu'))

# Camada de Saída
regressor.add(Dense(units=1, activation='linear'))
regressor.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mean_absolute_error'])

regressor.fit(data, price_target, batch_size = 300, epochs=100)


```

```python
predicitions = regressor.predict(data)
```

## Validação Cruzada

```python
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
import keras
```

```python
def build_neural_network():
    regressor = Sequential()
    regressor.add(Dense(units=158, activation='relu',input_dim=323))
    regressor.add(Dense(units=158, activation='relu'))
    regressor.add(Dense(units=1, activation='linear'))
    regressor.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mean_absolute_error'])
    
    return regressor
```
